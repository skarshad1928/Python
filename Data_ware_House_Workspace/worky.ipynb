```python

param_grid = [
    {
        'solver': ['liblinear'],
        'penalty': ['l1', 'l2'],
        'C': [0.01, 0.1, 1, 10, 100]
    },
    {
        'solver': ['lbfgs'],
        'penalty': ['l2', None],
        'C': [0.01, 0.1, 1, 10, 100]
    },
    {
        'solver': ['saga'],
        'penalty': ['l1', 'l2', 'elasticnet', None],
        'C': [0.01, 0.1, 1, 10, 100],
        'l1_ratio': [0, 0.5, 1]
    }
]


```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
import warnings 

warnings.filterwarnings("ignore")

log_reg = LogisticRegression(max_iter=1000)

grid = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,  
    cv=3,                   
    scoring='accuracy',
    n_jobs=-1               
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

```python
best_model = grid.best_estimator_

y_pred = best_model.predict(X_test)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

y_prob = best_model.predict_proba(X_test)[:, 1]
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label="ROC curve")
plt.plot([0,1],[0,1],'--',color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python

import numpy as np
import pandas as pd

best_model = grid.best_estimator_   
intercept = best_model.intercept_[0]    
coefs = best_model.coef_[0]             
if isinstance(X_train, pd.DataFrame):
    feature_names = X_train.columns.tolist()
else:
    n_features = coefs.shape[0]
    feature_names = [f"x{i+1}" for i in range(n_features)]

terms = [f"({intercept:.4f})"]
for name, c in zip(feature_names, coefs):
    terms.append(f"({c:.4f} * {name})")
linear_eq = " + ".join(terms)
print("Linear equation (logit):")
print("z = " + linear_eq)

print("\nProbability P(y=1|X) = 1 / (1 + exp(-z))")
print("So:")
print("P(y=1|X) = 1 / (1 + exp(- ( {} )))".format(linear_eq))

print("\nReadable form:")
readable = f"z = {intercept:.4f}"
for name, c in zip(feature_names, coefs):
    sign = "+" if c >= 0 else "-"
    readable += f" {sign} {abs(c):.4f}*{name}"
print(readable)
print("\nPredict probability example for a single sample X0 (array or Series):")
print("prob = 1 / (1 + np.exp(- ( intercept + sum(coef * x0) ) ))")

def predict_prob_from_vector(x):
    """
    x: 1D array-like or pandas Series with same order as X_train columns
    returns probability of class 1
    """
    x_arr = np.asarray(x)
    z = intercept + np.dot(coefs, x_arr)
    prob = 1.0 / (1.0 + np.exp(-z))
    return prob

example_row = X_test.iloc[0] if isinstance(X_test, pd.DataFrame) else X_test[0]
print("\nExample predicted probability (first test sample):",
      predict_prob_from_vector(example_row))

odds_ratio = np.exp(coefs)
print("\nOdds ratios (exp(coef)) per feature:")
for name, orr in zip(feature_names, odds_ratio):
    print(f"{name}: {orr:.4f}")
```

```python
import math

def churn_probability():
    intercept = -0.0282
    coefs = {
        "Gender": -0.0385,
        "Senior_Citizen": 0.2056,
        "Partner": 0.2734,
        "Dependents": -1.5246,
        "Tenure_Months": -0.0441,
        "Phone_Service": -0.7394,
        "Multiple_Lines": 0.5723,
        "Internet_Service": -0.3846,
        "Online_Security": -0.1107,
        "Online_Backup": 0.2146,
        "Device_Protection": 0.3657,
        "Tech_Support": -0.0932,
        "Streaming_TV": 0.6881,
        "Contract": -0.8864,
        "Paperless_Billing": 0.4989,
        "Payment_Method": 0.0499,
        "CLTV": 0.0000
    }

    values = {}
    for feature in coefs.keys():
        values[feature] = float(input(f"Enter {feature}: "))

    z = intercept + sum(coefs[f] * values[f] for f in coefs)

    p = 1 / (1 + math.exp(-z))

    print("\nPredicted Probability (P(y=1|X)):", round(p, 4))

    label = 1 if p > 0.328 else 0
    print("Predicted Class (with threshold 0.328):", label)


churn_probability()
```
